```python
# percentage score per students
df["percent"] = round((df["total"] / 300) * 100)
df
```

```python
def grade(score):
    if score >= 90:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 50:
        return "C"
    else:
        return "F"
```

```python
df["grade"] = df["percent"].apply(grade)
df
```
